```python
import numpy as np

# Zadane točke
points = np.array([[0, 6], [1, 0], [2, 0]])

# Definiranje matrice dizajna A
A = np.vstack([np.ones(len(points)), points[:, 0]]).T

# Vektor odstupanja b
b = points[:, 1]

# Izračun pseudoinverza matrice A
A_inv = np.linalg.pinv(A)

# Izračun vektora parametara x
x = np.dot(A_inv, b)

# Parametri pravca
C, D = x

print("Parametri pravca:")
print("C =", C)
print("D =", D)

# Ispis jednadžbe pravca
print("Jednadžba pravca: y =", C, "+", D, "x")
```

```python
#točke (0,6), (1,0), (2,0)
#C+Dt
#C+D*0=6
#C+D*1=0
#C+D*2=0
#Ax=b
#početna matrica
A = [[1,0],[1,1],[1,2]]
#y-vrijednost točaka
b = [6,0,0]
#transponirana matrica A
At=np.transpose(A)
# Matrično množenje (AtA)\(Atb) u pythonu ne može ovako
#x1 = (AtA)
x1 = np.matmul(At, A)
#x2 = (Atb)
x2 = np.matmul(At, b)
#x = (AtA)\(Atb)
x = np.linalg.solve(x1,x2)
#prvi broj je C, drugi je D
print(x)
```

```python
#točke (-2,0),(-1,0),(0,1),(1,0),(2,0)
#C+Dt+Et^2
#C+D*-2 + E*4=0
#C+D*-1 + E*1=0
#C+D*0 + E*0=1
#C+D*1 + E*1=0
#C+D*2 + E*4=0
#Ax=b
#početna matrica
A = [[1,-2,4],[1,-1,1],[1,0,0],[1,1,1], [1,2,4]]
#y-vrijednost točaka
b = [0,0,1,0,0]
#transponirana matrica A
At=np.transpose(A)
# Matrično množenje (AtA)\(Atb) u pythonu ne može ovako
#x1 = (AtA)
x1 = np.matmul(At, A)
#x2 = (Atb)
x2 = np.matmul(At, b)
#x = (AtA)\(Atb)
x = np.linalg.solve(x1,x2)
#prvi broj je C, drugi je D
print(x)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Matrica A i vektor b
A = np.array([[1, -2, 4], [1, -1, 1], [1, 0, 0], [1, 1, 1], [1, 2, 4]])
b = np.array([0, 0, 1, 0, 0])

# Rješavanje sustava Ax = b
x = np.linalg.lstsq(A, b, rcond=None)[0]

# Funkcija za izračun y vrijednosti koristeći dobivene parametre
def y_value(x_val):
    return x[0] + x[1]*x_val + x[2]*x_val**2

# Generiranje x vrijednosti
x_values = np.linspace(-2, 2, 100)
# Izračunavanje odgovarajućih y vrijednosti
y_values = y_value(x_values)

# Crtanje točaka
plt.plot([-2, -1, 0, 1, 2], [0, 0, 1, 0, 0], 'ro', label='Točke')
# Crtanje krivulje koja se aproksimira
plt.plot(x_values, y_values, 'b-', label='Krivulja aproksimacije')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Aproksimacija pomoću metode najmanjih kvadrata')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Matrica A i vektor b
A = np.array([[1, -2, 4], [1, -1, 1], [1, 0, 0], [1, 1, 1], [1, 2, 4]])
b = np.array([0, 0, 1, 0, 0])

# Rješavanje sustava Ax = b
x = np.linalg.lstsq(A, b, rcond=None)[0]

# Funkcija za izračun y vrijednosti koristeći dobivene parametre
def y_value(x_val):
    return x[0] + x[1]*x_val + x[2]*x_val**2

# Generiranje x vrijednosti
x_values = np.linspace(-2, 2, 100)
# Izračunavanje odgovarajućih y vrijednosti
y_values = y_value(x_values)

# Izračunavanje greške
errors = np.abs(b - A.dot(x))

# Crtanje točaka
plt.plot([-2, -1, 0, 1, 2], [0, 0, 1, 0, 0], 'ro', label='Točke')
# Crtanje krivulje koja se aproksimira
plt.plot(x_values, y_values, 'b-', label='Krivulja aproksimacije')
# Crtanje greške iznad svake točke
plt.errorbar([-2, -1, 0, 1, 2], b, yerr=errors, fmt='none', ecolor='gray', capsize=5, label='Greška')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Aproksimacija pomoću metode najmanjih kvadrata s greškom')
plt.legend()
plt.grid(True)
plt.show()
```
